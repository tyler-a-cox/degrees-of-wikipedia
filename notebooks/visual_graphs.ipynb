```python
import numpy as np
import pylab as plt
import networkx as nx
from copy import deepcopy
from wikiracer.utils import load_graph
plt.rcParams.update({"text.usetex": False})
```

```python
graph = load_graph('../graph.pkl')
```

```python
total_links = 0
links = set()

for key in graph.keys():
    print (key, len(graph[key]))
    links.update(graph[key])
    total_links += len(graph[key])
```

```python
def prune_graph(graph, nodes=['Turmeric', '13-Hydroxyoctadecadienoic acid', 'Rhizome']):
    """
    """
    pruned = deepcopy(graph)
    for node in nodes:
        for value in graph[node]:
            if pruned.get(value) is None:
                pruned[node].remove(value)
                
    return pruned
```

```python
pruned = prune_graph(graph)
edges = []
for key, values in pruned.items():
    edges += [(key, value) for value in values]
    
edges = list(set(edges))
```

```python
g = nx.Graph()
g.add_edges_from(edges)
```

```python
plt.figure(figsize=(10, 10))
nx.draw_networkx(g, with_labels=False)
plt.show()
```
